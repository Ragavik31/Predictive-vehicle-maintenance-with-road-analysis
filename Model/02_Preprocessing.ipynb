```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("data/raw/logistics_predictive_maintenanceV2.csv")

df.head()
```

```python
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
```

```python
df["Last_Maintenance_Date"] = pd.to_datetime(
    df["Last_Maintenance_Date"]
)
```

```python
numerical_columns = df.select_dtypes(
    include=["int64","float64"]
).columns.tolist()

categorical_columns = df.select_dtypes(
    include="object"
).columns.tolist()

print("Numerical:", len(numerical_columns))
print("Categorical:", len(categorical_columns))
```

```python
label_encoders = {}

for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le
```

```python
df["Maintenance_Required"].value_counts()
df["Maintenance_Required"].value_counts(normalize=True) * 100
```

```python
df.to_csv(
    "data/processed/preprocessed_dataset.csv",
    index=False
)
```
